# Introduction to Julia Language

Julia is a high-level and high-performance programming language that has gained popularity in recent years. It's designed to be as easy to learn as Python while still giving users the speed of C/C++! Right now, it's mainly being used in Data Science and Technical Computing, but many people believe it will overtake Python in the near future. 

Here's a bit of background on Julia and what makes it so exceptional.

## Why was Julia created?

In the original blog post by Bezanson, Karpinski, Shah, and Edelman, they state the following: 

"*In short, because we are greedy.*

*We want a language that's open source, with a liberal license. We want the speed of C with the dynamism of Ruby. We want a language that's homoiconic, with true macros like Lisp, but with obvious, familiar mathematical notation like Matlab. We want something as usable for general programming as Python, as easy for statistics as R, as natural for string processing as Perl, as powerful for linear algebra as Matlab, as good at gluing programs together as the shell. Something that is dirt simple to learn, yet keeps the most serious hackers happy. We want it interactive and we want it compiled.*‚Äù

The creators claim it does 90% of the above.

### Claim 3: Dynamism of Ruby

Dynamic languages are generally ‚Äúhigh-level‚Äù languages that perform certain behaviors at runtime rather than at compile time. One common association with dynamic languages is dynamic ‚Äútyping‚Äù.

Python, Ruby, JavaScript, PHP, and Perl are some of the most popular dynamic languages.

Most dynamic languages are interpreted rather than compiled. This is usually what allows for dynamic typing. This includes Python, Ruby, JavaScript, PHP, and Perl (although Perl uses a unique interpreter).

**This is where Julia is unique.** Julia is a compiled language that uses a unique version of **Just-In-Time Compilation**. It only compiles code that is strictly necessary. This allows it to have the dynamism of similar interpreted languages without sacrificing speed.

So ultimately, this does deliver! Julia is also a dynamic language.


### Claim 4: Homoiconic, with true macros like Lisp

Homoiconic is usually defined as the following: A program written in this language can be maniuplated as data using this language.

An example of this is what's known as first-class functions, or functions that can be passed as arguments to other functions.

There is some controversy over what counts as a homoiconic language. The vast majority of languages are **able** to be written in such a way that this is true. But most people argue that for a language to be considered homoiconic, it must lend itself to this easily. 

Kent Pitman, a programmer hugely involved in the design and implementation of List and many Lisp-based systems, said the following:

‚Äú*I like Lisp's willingness to represent itself. People often explain this as its ability to represent itself, but I think that's wrong. **Most languages are capable of representing themselves, but they simply don't have the will to.** Lisp programs are represented by lists and programmers are aware of that. It wouldn't matter if it had been arrays. It does matter that it's program structure that is represented, and not character syntax, but beyond that the choice is pretty arbitrary. It's not important that the representation be the Right¬Æ choice. It's just important that it be a common, agreed-upon choice so that there can be a rich community of program-manipulating programs that "do trade" in this common representation.*‚Äù

While you can find many passionate arguments going both ways, the general consensus is that Julia is **not** a homoiconic language. While it has the ability to be, writing Julia functions in a homoiconic way requires you to re-imagine the logic of typical Julia code. So unfortunately, this claim does not hold up.

### Claim 8: As natural for string processing as Perl

Perl has a powerful regular expression engine, hence why it is considered natural for string processing.
Julia has excellent support for RegEx as well. 
It also has the advantage of being compiled vs. interpreted- this helps with speed.
Some speed benchmarks even put Julia above Perl for String Processing.

Perl also has the disadvantage of being less readable and harder to learn than Julia.

So this claim holds up!
